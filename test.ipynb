```python
import pyttsx3
import time
```

```python
engine = pyttsx3.init("sapi5")
voices = engine.getProperty("voices")
engine.setProperty("rate", 170)
engine.setProperty("voice", voices[1].id)
```

```python
print(voices)
```

```python
print(voices[0].id)
```

```python
print(voices[1].id)
```

```python
def speak(text):
    """
    This function is used to speak the text
    Args:
        text (str): The text to speak
    Returns:
      voice
    """
    engine.say(text)
    engine.runAndWait()
   
```

```python
speak("Hello, how are you?")
time.sleep(1)
speak("I am fine,My name is Prodip Sarkar")
```
